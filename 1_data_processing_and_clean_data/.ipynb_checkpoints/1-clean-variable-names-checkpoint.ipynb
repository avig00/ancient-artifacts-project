```python
# import required package

import pandas as pd
import numpy as np
```

```python
arch_soil = pd.read_csv("../0_raw_data/ArchaeologicalSoilData.csv")
lithic_exp = pd.read_csv("../0_raw_data/LithicExperimentalData.csv")
```

```python
col_with_unit = [i for i in range(len(arch_soil.iloc[0,])) if "mm" in arch_soil.iloc[0,i]]
replace_col_name = [i + "(" + j + ")" for i,j in zip(arch_soil.columns, arch_soil.iloc[0,]) if "mm" in j]

arch_soil.columns.values[col_with_unit] = replace_col_name
```

```python
new_arch_soil = arch_soil[1:]
new_arch_soil
```

```python
lithic_exp.columns.values[col_with_unit] = replace_col_name
```

```python
new_lithic_exp = lithic_exp[1:]
new_lithic_exp
```

```python
new_arch_soil.to_csv("new_arch_soil")
new_lithic_exp.to_csv("new_lithic_exp")
```
