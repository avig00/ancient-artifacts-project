# EDA

## Types

```python
micro.info()
```

### Drop features start with "filter"

```python
# Drop the columns (filter1, filter2, ...)
to_drop = micro.filter(regex='^Filter').columns
micro = micro.drop(columns=to_drop)
micro
```

### Drop features: "hash", "Krumbein Rnd"

```python
micro = micro.drop(columns=['hash', 'Krumbein Rnd']) #'Krumbein Rnd'
micro
```

#### Drop id and image id

```python
micro = micro.iloc[: , 2:]
micro
```

## Descriptives

```python
micro.describe()
```

## Visualizations

### Size

Compare means

```python
size_dimensions = ['Da', 'Dp', 'FLength', 'FWidth', 'FThickness', 'ELength', 'EThickness', 'EWidth', 
               'type_micro']
size_whole = ['Volume', 'Area', 'Surface Area', 'CHull  Area', 'CHull Surface Area', 'Perimeter',
              'type_micro']
size_other = ['CHull Perimeter', 'Sphericity', 'Sieve','Fiber Length','Fiber Width', 
              'type_micro']

```

```python
mean_by_group1 = micro[size_dimensions].groupby('type_micro').mean()
```

```python
mean_by_group1.T.plot(kind='bar', stacked=False)
plt.xticks(rotation=45)
```

```python
mean_by_group2 = micro[size_whole].groupby('type_micro').mean()
```

```python
mean_by_group2.T.plot(kind='bar', stacked=False)
plt.xticks(rotation=45)
```

```python
mean_by_group3 = micro[size_other].groupby('type_micro').mean()
```

```python
mean_by_group3.T.plot(kind='bar', stacked=False)
plt.xticks(rotation=45)
```

correlation matrix

```python
corr_micro = micro[size_dimensions + size_whole + size_other]
corr_matrix = corr_micro.corr()
```

```python
sns.heatmap(corr_matrix, cmap="coolwarm")
```

```python
corr_matrix.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
micro.head()
```

### Shape/Form

Compare means

```python
shape = ['Sphericity', 'Circularity', 'Roundness', 'Extent', 'Ellipticity', 'Compactness', 
               'type_micro']
shape_angularity = ['Angularity', 'type_micro']
shape_ratio = ['Ellipse Ratio', 'W/L Ratio', 'L/W Ratio', 'T/L Aspect Ratio', 'L/T Ratio', 'T/W Ratio', 'W/T Ratio',
              'type_micro']

```

```python
mean_by_group1 = micro[shape].groupby('type_micro').mean()
mean_by_group1.T.plot(kind='bar', stacked=False)
plt.xticks(rotation=45)
```

```python
mean_by_group = micro[shape_angularity].groupby('type_micro').mean()
mean_by_group.T.plot(kind='bar', stacked=False)
plt.xticks(rotation=45)
```

```python
mean_by_group2 = micro[shape_ratio].groupby('type_micro').mean()
mean_by_group2.T.plot(kind='bar', stacked=False)
plt.xticks(rotation=45)
```

correlation matrix

```python
corr_micro = micro[shape + shape_ratio]
corr_matrix = corr_micro.corr()
```

```python
sns.heatmap(corr_matrix, cmap="coolwarm")
```

```python
corr_matrix.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
micro.head()
```
