### Dealing with Sample Imbalance

Oversampling: This technique involves generating new samples for the minority class by duplicating existing samples. This will increase the number of samples for the minority class, making the dataset more balanced.

Undersampling: This technique involves removing samples from the majority class to make the dataset more balanced. This can be done by randomly removing samples or by using a more sophisticated method such as Tomek links or SMOTE

Cost-sensitive learning: This technique involves adjusting the cost of misclassifying samples from different classes in the learning algorithm. This can be done by adjusting the weights of the samples or by adjusting the decision threshold of the model.

Ensemble methods: Ensemble methods such as bagging and boosting can be used to create multiple models that are combined to make a final prediction. These methods can be particularly useful when dealing with class imbalance because they can help to reduce the impact of the imbalance on the final model.

Changing the performance metric: Instead of using accuracy as a performance metric, you can use F1-score, precision, recall, or area under the receiver operating characteristic (ROC) curve, which are more robust to class imbalance.

#### Undersampling

```python
micro.groupby('type_micro').describe()
```

```python
# Separate majority and minority classes
df_majority = micro[micro.type_micro=='soil']
df_minority = micro[micro.type_micro=='lithic']

# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=5299,     # to match minority class
                                 random_state=123) # reproducible results
 
# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
```

```python
df_downsampled
```

```python
# Split the data into features (X) and labels (y)
X = df_downsampled.drop("type_micro", axis=1)
y = df_downsampled["type_micro"]

# Split the data into training and test sets (80% training, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create a random forest classification model
model = RandomForestClassifier(n_estimators=100)

# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Access the feature importances
importances = model.feature_importances_

# Print the feature importances
for feature, importance in zip(X.columns, importances):
    print(f'{feature} : {importance}')

# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange feature names so they match the sorted feature importances
names = [X.columns[i] for i in indices]

# Create plot
plt.bar(range(X.shape[1]), importances[indices])

# Add feature names as x-axis labels
plt.xticks(range(X.shape[1]), names, rotation=90)

# Show plot
plt.show()
```

```python
cm = confusion_matrix(y_test, y_pred)
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

sns.heatmap(cm_norm, annot=True, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Normalized Confusion Matrix')
plt.show()
```
