```python
df = {'model': ['LightGBM', 'AdaBoost', 'Random Forest', 'Bagging', 'Extra Trees', 'SVC', 'Decision Tree'],
            'performance': [0.94, 0.92, 0.98, 0.91, 0.93, 0.96, 0.92],
            'time taken': [0.28, 1.36, 1.99, 1.09, 0.72, 1.6,0.21]}

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(df['time taken'], df['performance'], color='royalblue', s=50)
for i, txt in enumerate(df['model']):
    ax.annotate(txt, (df['time taken'][i] + 0.04, df['performance'][i] + 0.002))
ax.set_xlabel('Time Taken (seconds)')
ax.set_ylabel('Performance (Accuracy)')
ax.set_title('Performance vs. Time Taken for Best Models')
ax.spines[['right', 'top']].set_visible(False)
ax.set_ylim([0.85, 1])
plt.show()
```
